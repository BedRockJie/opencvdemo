## 截取一部分图像

```python
img = cv2.imread('Nemesis.jpg')
cat = img[0:200,0:200]
cvshow('cat',cat)
```

```python
b,g,r = cv2.split(img)
```

```python
b.shape
```

```python
img = cv2.merge((b,g,r)) #把他们组合进去
img.shape
```

```python
#只保留R
cur_img = img.copy()
cur_img[:,:,0] = 0
cur_img[:,:,1] = 0
cvshow('R',cur_img)
```

```python
#只保留B
cur_img = img.copy()
cur_img[:,:,1] = 0
cur_img[:,:,2] = 0
cvshow('B',cur_img)
```

```python
#只保留G
cur_img = img.copy()
cur_img[:,:,0] = 0
cur_img[:,:,2] = 0
cvshow('B',cur_img)
```
