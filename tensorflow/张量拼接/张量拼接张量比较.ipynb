# 数据统计
向量范数时使用表征向量长度的方法  表示向量的全职大小 和 梯度大小
- L1范数  定义为向量的所有元素的权值之和
- L2范数  定义为向量所有元素的平方和再开根号
- $\infty$范数 定义为元素绝对值的最大值

### TF中 MSE 误差函数可以求出每个样本的误差,需要计算样本的平均误差,使用tf.reduce_mean在样本维度上计算均值:

```python
from tensorflow import keras
out = tf.random.normal([4,10]) #模拟预测输出
y = tf.constant([1,2,2,0]) #模拟真实标签
y = tf.one_hot(y,depth = 10) #one_hot编码
loss = keras.losses.mse(y,out) #计算每个样本的误差
loss = tf.reduce_mean(loss) #在每个样本上取均值
loss #误差标量
```

```python
tf.reduce_sum(out, axis = -1) #求最后一个维度的和
```

```python
out= tf.nn.softmax(out,axis=1)#通过softmax函数转化 为概率值
out
```

```python
pred = tf.argmax(out,axis = 1) #选取概率最大的位置
pred
```

## 张量比较

```python
out = tf.random.normal([100,10])
out = tf.nn.softmax(out,axis=1)#输出转化为概率
pred = tf.argmax(out,axis=1)#计算预测值
pred
```

```python
y = tf.random.uniform([100],dtype = tf.int64,maxval=10)#生成模型的真实标签
y
```

```python
out = tf.equal(pred,y) #预测值与真实值比较，返回布尔类型的张量
out
```

```python
out = tf.cast(out , dtype = tf.float32)  # 布尔型转化为 int 型 
correct = tf.reduce_sum(out) #统计True 的个数
correct   #因此模型预测的正确个数是10% 这也是正常水平 
```

```python
#tf.meshgrid 函数可以方便地生成二维网格的采样点坐标，方便可视化等应用场合。
# from math import sinc
import numpy as np 
points = []
for x in range(-8,8,100):
    for y in range(-8,8 ,100):
        z = np.sinc(x,y)
        points.append([x,y,z])
```

```python
x = tf.linspace(-8.,8,100)
y = tf.linspace(-8.,8,100)
x,y = tf.meshgrid(x,y)
x.shape,y.shape
```

```python
z = tf.sqrt(x**2 + y**2)
z = tf.sin(z) / 2      # sinc 函数
import matplotlib
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig) # 设置 3D 坐标轴
# 根据网格点绘制 sinc 函数 3D 曲面
ax.contour3D(x.numpy(),y.numpy(),z.numpy(),50)
plt.show()
```
