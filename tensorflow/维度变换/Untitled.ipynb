#### 创建
 tf.random.normal(shape, mean=0.0, stddev=1.0)可以创建形状为 shape，均值为
mean，标准差为 stddev 的正态分布𝒩

```python
a = tf.random.normal([4,28,28,3]) #创建正态分布的张量
a
```

```python
a.shape,a.ndim
```

```python
tf.reshape(a,[4,28*28,3]).shape
```

```python
tf.reshape(a,[4,-1,3]).shape
```

```python
tf.reshape(a,[4,784*3]).shape
```

```python
tf.reshape(a,[4,-1]).shape
```

```python
a.shape,a.ndim
```

### 矩阵的转置
[w,h]   [h,w]

```python
a = tf.random.normal((4,3,2,1))
```

```python
a.shape
```

```python
tf.transpose(a).shape
```

```python
tf.transpose(a,perm=[0,1,3,2]).shape
```

## 各种维度

- [b,h,w,c]  tensor
- [b,c,h,w]  使用[0,3,1,2]来进行交换  ptouch所接受

```python
#增加一个维度 dim
a = tf.random.normal([4,35,8])
```

```python
a = tf.expand_dims(a,axis=0).shape
a
```

```python
#根据index 不同会在不同地方增加一个维度
```

```python
#减少一个维度，只能较少维度为1的shape
tf.squeeze(a)
```

#### Broadcasting 
在某一个维度上  复制n多次（扩张但不复制数据）  

相比于 tf.tile并不真正扩张 

```python
#交换维度
x = tf.random.normal([2,32,32,3])
tf.transpose(x,perm=[0,3,1,2])    #即矩阵的转置
```
